### Importa칞칚o dos dados



```python
# Importa칞칚o das bibliotecas utilizadas

import pandas as pd  # Biblioteca para manipula칞ao e an치lise de dados.
import matplotlib.pyplot as plt #Biblioteca para cria칞칚o de gr치fico (2D).
import seaborn as sns ## Biblioteca baseada no Matplotlib, que fornece uma interface mais simples para gr치ficos estat칤sticos.
import plotly.io as pio
pio.renderers.default = 'colab'

#importa칞칚o dos dadosno github

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

#Transformando os arquivos CVS em Dataframes e lendo os arquivos com a fun칞칚o pd.read_csv()
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Verificando as 5 primeiras linhas do Dataframe loja.
loja.head()
```

#1. An치lise do faturamento


```python
#Somando os valores da coluna Pre칞o de cada loja para saber o faturamento total de cada loja.

faturamento_loja1 = loja['Pre칞o'].sum()
faturamento_loja2 = loja2['Pre칞o'].sum()
faturamento_loja3 = loja3['Pre칞o'].sum()
faturamento_loja4 = loja4['Pre칞o'].sum()

print("游눯 Faturamento da loja 1: R$", round(faturamento_loja1, 2))
print("游눯 Faturamento da loja 2: R$", round(faturamento_loja2, 2))
print("游눯 Faturamento da loja 3: R$", round(faturamento_loja3, 2))
print("游눯 Faturamento da loja 4: R$", round(faturamento_loja4, 2))
```

```python
#Tratando os dados para fazer o gr치fico e comparar o faturamento de cada loja.

faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamento, color =['blue', 'green','orange','red'])
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.title('Faturamento das Lojas')

# Adicionar os valores nas barras
for i, valor in enumerate(faturamento):
    plt.text(i, valor + 1000, f'R$ {valor:,.0f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```

# 3. M칠dia de Avalia칞칚o das Lojas

# 4. Produtos Mais e Menos Vendidos

```python
# Fun칞칚o para gerar o ranking por loja
def mais_menos_vendidos(df, nome_loja):
    grupo = df.groupby("Produto")["Avalia칞칚o da compra"].sum()
    mais_vendido = grupo.idxmax()
    menos_vendido = grupo.idxmin()
    print(f"\n{' Resultados - ' + nome_loja + ' ':=^50}")
    print(f"游늷 Loja: {nome_loja}")
    print(f"游댠 Produto mais vendido: {mais_vendido} ({grupo[mais_vendido]} vendas)")
    print(f"仇勇뀷roduto menos vendido: {menos_vendido} ({grupo[menos_vendido]} vendas)")

    return mais_vendido, menos_vendido #Retorna os valores

#Armazena os valores em dicion치rios
mais_vendidos = {}
menos_vendidos = {}

# Chama a fun칞칚o para cada loja e armazena os resultados
mais_vendidos["Loja 1"], menos_vendidos["Loja 1"] = mais_menos_vendidos(loja, "Loja 1")
mais_vendidos["Loja 2"], menos_vendidos["Loja 2"] = mais_menos_vendidos(loja2, "Loja 2")
mais_vendidos["Loja 3"], menos_vendidos["Loja 3"] = mais_menos_vendidos(loja3, "Loja 3")
mais_vendidos["Loja 4"], menos_vendidos["Loja 4"] = mais_menos_vendidos(loja4, "Loja 4")
```
