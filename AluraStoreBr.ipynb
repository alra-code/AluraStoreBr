### Importação dos dados



```python
# Importação das bibliotecas utilizadas

import pandas as pd  # Biblioteca para manipulaçao e análise de dados.
import matplotlib.pyplot as plt #Biblioteca para criação de gráfico (2D).
import seaborn as sns ## Biblioteca baseada no Matplotlib, que fornece uma interface mais simples para gráficos estatísticos.
import plotly.io as pio
pio.renderers.default = 'colab'

#importação dos dadosno github

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

#Transformando os arquivos CVS em Dataframes e lendo os arquivos com a função pd.read_csv()
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Verificando as 5 primeiras linhas do Dataframe loja.
loja.head()
```

#1. Análise do faturamento


```python
#Somando os valores da coluna Preço de cada loja para saber o faturamento total de cada loja.

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("💰 Faturamento da loja 1: R$", round(faturamento_loja1, 2))
print("💰 Faturamento da loja 2: R$", round(faturamento_loja2, 2))
print("💰 Faturamento da loja 3: R$", round(faturamento_loja3, 2))
print("💰 Faturamento da loja 4: R$", round(faturamento_loja4, 2))
```

```python
#Tratando os dados para fazer o gráfico e comparar o faturamento de cada loja.

faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamento, color =['blue', 'green','orange','red'])
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.title('Faturamento das Lojas')

# Adicionar os valores nas barras
for i, valor in enumerate(faturamento):
    plt.text(i, valor + 1000, f'R$ {valor:,.0f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```

# 2. Vendas por Categoria


```python
#Somando as vendas por categoria de cada loja
categorias_loja1 = loja['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()

print("📊 Vendas por Categoria da loja 1:\n", categorias_loja1, '\n')
print("📊 Vendas por Categoria da loja 2:\n", categorias_loja2, '\n')
print("📊 Vendas por Categoria da loja 3:\n", categorias_loja3, '\n')
print("📊 Vendas por Categoria da loja 4:\n", categorias_loja4, '\n')
```

```python
#Utilizando gráficos para visualizar melhor os resultados
plt.figure(figsize=(8, 5))
plt.barh(categorias_loja1.index, categorias_loja1.values, color='blue')
plt.title('Categorias Mais Vendidas - Loja 1')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja2.index, categorias_loja2.values, color='green')
plt.title('Categorias Mais Vendidas - Loja 2')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja3.index, categorias_loja3.values, color='orange')
plt.title('Categorias Mais Vendidas - Loja 3')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja4.index, categorias_loja4.values, color='red')
plt.title('Categorias Mais Vendidas - Loja 4')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
#Calculndo a média das avaliações
med_avaliacao_loja1 = loja['Avaliação da compra'].mean()
med_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
med_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
med_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print("⭐ Média de Avaliação das Lojas 1:", med_avaliacao_loja1.round(2))
print("⭐ Média de Avaliação das Lojas 2:", med_avaliacao_loja2.round(2))
print("⭐ Média de Avaliação das Lojas 3:", med_avaliacao_loja3.round(2))
print("⭐ Média de Avaliação das Lojas 4:", med_avaliacao_loja4.round(2))

```

```python
#Tratando os dados para fazer o gráfico e comparar o faturamento de cada loja.
import matplotlib.pyplot as plt
media_avaliacao = [med_avaliacao_loja1, med_avaliacao_loja2, med_avaliacao_loja3, med_avaliacao_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
cores = ['blue', 'green', 'orange', 'red']

#Gerando gráfico da média de avaliações por loja
for i in range(len(lojas)):
  plt.plot(lojas[i], media_avaliacao[i], marker='o', markerfacecolor=cores[i])
plt.title('Média de Avaliação das Lojas')
plt.xlabel('')
plt.ylabel('Avaliação')
plt.grid(True)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para gerar o ranking por loja
def mais_menos_vendidos(df, nome_loja):
    grupo = df.groupby("Produto")["Avaliação da compra"].sum()
    mais_vendido = grupo.idxmax()
    menos_vendido = grupo.idxmin()
    print(f"\n{' Resultados - ' + nome_loja + ' ':=^50}")
    print(f"📌 Loja: {nome_loja}")
    print(f"🔥 Produto mais vendido: {mais_vendido} ({grupo[mais_vendido]} vendas)")
    print(f"❄️Produto menos vendido: {menos_vendido} ({grupo[menos_vendido]} vendas)")

    return mais_vendido, menos_vendido #Retorna os valores

#Armazena os valores em dicionários
mais_vendidos = {}
menos_vendidos = {}

# Chama a função para cada loja e armazena os resultados
mais_vendidos["Loja 1"], menos_vendidos["Loja 1"] = mais_menos_vendidos(loja, "Loja 1")
mais_vendidos["Loja 2"], menos_vendidos["Loja 2"] = mais_menos_vendidos(loja2, "Loja 2")
mais_vendidos["Loja 3"], menos_vendidos["Loja 3"] = mais_menos_vendidos(loja3, "Loja 3")
mais_vendidos["Loja 4"], menos_vendidos["Loja 4"] = mais_menos_vendidos(loja4, "Loja 4")
```

```python
#Produtos mais vendidos em todas as lojas
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dados das lojas e formar um único Dataframe
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Concatenando os DataFrames
todos_as_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Contar as ocorrências de cada produto
contagem = todos_as_lojas['Produto'].value_counts(ascending=False)
contagem = contagem.head(10)
print(contagem)
```

```python
# Cria o gráfico de barras invertidas
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.barh(contagem.index, contagem.values, color='blue')
plt.title('10 Produtos Mais Vendidos (Todas as Lojas)')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Produto')
plt.gca().invert_yaxis()  # Inverte o eixo y para exibir as barras na ordem decrescente
plt.tight_layout()  # Ajusta o layout para evitar sobreposição de elementos
plt.show()
```

# 5. Frete Médio por Loja


```python
#Calculndo o frete médio por loja
frete_medio_loja = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print("🚚 Frete médio da loja 1: R$", round(frete_medio_loja, 2))
print("🚚 Frete médio da loja 2: R$", round(frete_medio_loja2, 2))
print("🚚 Frete médio da loja 3: R$", round(frete_medio_loja3, 2))
print("🚚 Frete médio da loja 4: R$", round(frete_medio_loja4, 2))
```

```python
#Tratando os dados para fazer o gráfico e comparar o Frete médio de cada loja.

frete_medio = [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
plt.bar(lojas, frete_medio, color =['blue', 'green','orange','red'])
plt.xlabel('Lojas')
plt.ylabel('Frete')
plt.title('Frete médio das Lojas')
plt.show()
```
